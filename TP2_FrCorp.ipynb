## Part1 : NLTK

```python
import nltk
french_corp="""Il s’agit de la deuxième visite de Volodymyr Zelensky à Paris, depuis le début de l’offensive russe en Ukraine. "La dernière remonte au 8 février dernier, en présence du chancelier allemand", rappelle la journaliste. "Cette fois-ci, Emmanuel Macron et Volodymyr Zelensky seront en tête-à-tête, ils dîneront ce soir ensemble. Une visite éclair, car le président ukrainien repart dès demain matin", conclut Béatrice Gelot, en direct du palais de l’Elysée."""
french_corp
```

```python
sentence_tokenizing=nltk.sent_tokenize(french_corp)
print(sentence_tokenizing)
```

```python
tokens=nltk.word_tokenize(french_corp)
print(tokens)
```

```python
tagged = nltk.pos_tag(tokens)
tagged
```

```python
import svgling
entities = nltk.chunk.ne_chunk(tagged)
entities
```

## Part2 : Spacy

```python
import spacy
nlp = spacy.load("en_core_web_sm")
french_corp2=nlp("""Il s’agit de la deuxième visite de Volodymyr Zelensky à Paris, depuis le début de l’offensive russe en Ukraine. "La dernière remonte au 8 février dernier, en présence du chancelier allemand", rappelle la journaliste. "Cette fois-ci, Emmanuel Macron et Volodymyr Zelensky seront en tête-à-tête, ils dîneront ce soir ensemble. Une visite éclair, car le président ukrainien repart dès demain matin", conclut Béatrice Gelot, en direct du palais de l’Elysée.""")
print(french_corp2.text)
```

```python
import pandas as pd
spacy_taggs=pd.DataFrame(columns=['tokens'])

spacy_taggs
```

```python
for token in french_corp2:
    spacy_taggs = spacy_taggs.append({'tokens': token.text,
                    'lemma': token.lemma_,
                    'tokens_pos': token.pos_,
                    'tokens_tag': token.tag_,
                    'tokens_dep': token.dep_,
                    'tokens_shape': token.shape_,
                    'is_stop': token.is_stop}, ignore_index=True)
print(spacy_taggs)
```

```python
from spacy import displacy
displacy.render(french_corp2, style="dep", jupyter=True)
```

```python
for ent in french_corp2.ents:
    print(ent.text, ent.start_char, ent.end_char, ent.label_)
```

```python
from spacy import displacy
displacy.render(french_corp2, style="ent")
```

```python
import nltk
from nltk.corpus import stopwords
stop_words=set(stopwords.words("French"))

filtered_sentence=[]
for w in tokens:
    if w not in stop_words:
        filtered_sentence.append(w)
        
filtered_sentence
```

## Word Frequency

```python
freq=nltk.FreqDist(tokens)
freq
```

```python
print(freq.most_common(10))
```

```python
only_nn = [x for (x,y) in tagged if y in [('NNP'),('NNS'),('CD'),('JJ'),('VBN'),('VBD')]]
freq2=nltk.FreqDist(only_nn)
freq2
```

```python
print(freq2.most_common(10))
```

```python
from collections import Counter
tag_freq = Counter(tag for _, tag in tagged)
print(tag_freq)
```

```python
print(tag_freq.most_common(3))
```

## Bigrams 

```python
import nltk
nltk.download('punkt')
bg=nltk.bigrams(tokens)
f = nltk.FreqDist(bg)
for key,value in f.items():
    print(key,value)
```

```python
freq_list = [x for (x,y) in tagged if y in [('NNP'),('VBN'),('VBD')]]
print(freq_list)
```

```python
bg2=nltk.bigrams(freq_list)
f2 = nltk.FreqDist(bg2)
for key,value in f2.items():
    print(key,value)
```
