## Part1 : NLTK

```python
import nltk
arab_corp="عبد الرحمن بن محمد بن خلدون الحضرمي مؤرخ عربي أندلسي الأصل وعاش في أقطار شمال أفريقيا، رحل إلى بسكرة وفاس، وغرناطة، وبجاية وتلمسان، والأندلس، كما تَوَجَّهَ إلى مصر، حيث أكرمه سلطانها الظاهر برقوق، ووُلِّيَ فيها قضاء المالكية، وظلَّ بها ما يناهز ربع قرن (784-808 هـ)، حيث تُوُفِّيَ عام 1406 عن عمر بلغ ستة وسبعين عامًا ودُفِنَ قرب باب النصر بشمال القاهرة"
print(arab_corp)
```

```python
sentence_tokenizing=nltk.sent_tokenize(arab_corp)
print(sentence_tokenizing)
```

```python
tokens=nltk.word_tokenize(arab_corp)
print(tokens)
```

```python
tagged = nltk.pos_tag(tokens)
tagged
```

```python
import svgling
entities = nltk.chunk.ne_chunk(tagged)
entities
```

## Part2 : Spacy

```python
import spacy
nlp = spacy.load("en_core_web_sm")
arab_corp2=nlp("عبد الرحمن بن محمد بن خلدون الحضرمي مؤرخ عربي أندلسي الأصل وعاش في أقطار شمال أفريقيا، رحل إلى بسكرة وفاس، وغرناطة، وبجاية وتلمسان، والأندلس، كما تَوَجَّهَ إلى مصر، حيث أكرمه سلطانها الظاهر برقوق، ووُلِّيَ فيها قضاء المالكية، وظلَّ بها ما يناهز ربع قرن (784-808 هـ)، حيث تُوُفِّيَ عام 1406 عن عمر بلغ ستة وسبعين عامًا ودُفِنَ قرب باب النصر بشمال القاهرة")
print(arab_corp2.text)
```

```python
import pandas as pd
spacy_taggs=pd.DataFrame(columns=['tokens'])

spacy_taggs
```

```python
for token in arab_corp2:
    spacy_taggs = spacy_taggs.append({'tokens': token.text,
                    'lemma': token.lemma_,
                    'tokens_pos': token.pos_,
                    'tokens_tag': token.tag_,
                    'tokens_dep': token.dep_,
                    'tokens_shape': token.shape_,
                    'is_stop': token.is_stop}, ignore_index=True)
print(spacy_taggs)
```

```python
from spacy import displacy
displacy.render(arab_corp2, style="dep", jupyter=True)
```

```python
displacy.render(arab_corp2, style="ent")
```

## Word Frequency

```python
freq=nltk.FreqDist(tokens)
freq
```

```python
print(freq.most_common(10))
```

```python
only_nn = [x for (x,y) in tagged if y in [('NNP'),('NNS'),('CD'),('JJ'),('VBN'),('VBD')]]
freq2=nltk.FreqDist(only_nn)
freq2
```

```python
print(freq2.most_common(10))
```

```python
from collections import Counter
tag_freq = Counter(tag for _, tag in tagged)
print(tag_freq)
```

```python
print(tag_freq.most_common(3))
```

## Bigrams 

```python
import nltk
nltk.download('punkt')
bg=nltk.bigrams(tokens)
f = nltk.FreqDist(bg)
for key,value in f.items():
    print(key,value)
```

```python
freq_list = [x for (x,y) in tagged if y in [('NNP'),('VBN'),('VBD')]]
print(freq_list)
```

```python
bg2=nltk.bigrams(freq_list)
f2 = nltk.FreqDist(bg2)
for key,value in f2.items():
    print(key,value)
```
