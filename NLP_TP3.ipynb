```python
with open('Doc1_TP3.txt') as file :
    Doc1 = file.read()
    print(Doc1)
```

```python
import nltk
sentence_tokenizing=nltk.sent_tokenize(Doc1)
sentence_tokenizing
```

```python
tokens = nltk.word_tokenize(Doc1)
tokens
```

```python
nltk.download('stopwords')

from nltk.corpus import stopwords
stop_words=set(stopwords.words("English"))

filtered_Doc1=[]
for w in tokens:
    if w not in stop_words:
        filtered_Doc1.append(w)
        
#filtered_Doc1
```

```python
import pandas as pd

frequency=[]
count=0

for i in range(0,len(filtered_Doc1)):
    for j in range(0,len(filtered_Doc1)):
        if filtered_Doc1[i]==filtered_Doc1[j]:
           count=count+1
    frequency.append(count)
    
    #print(filtered_Doc1[i],count)
    count=0 

Bow_Bag = pd.DataFrame({'Filtered_Words' : filtered_Doc1 , 'Frequency': frequency})
Bow_Bag
```

```python
tf=[]
for i in range(0,len(filtered_Doc1)):
     tf.append(frequency[i]/len(filtered_Doc1))
        
Bow_Bag.insert(2,'TF',tf)

Bow_Bag
```

```python
tokens2=nltk.word_tokenize(Doc2)
print(tokens2)
```

```python
filtered_Doc2=[]
for w in tokens2:
    if w not in stop_words:
        filtered_Doc2.append(w)
        
#filtered_Doc2
```

```python
def ConvertToString(l):
 
    doc = ""
    for w in l:
        doc += w
 
    return doc
 
```

```python
n_Doc1=ConvertToString(filtered_Doc1)
n_Doc2=ConvertToString(filtered_Doc2)

n_Doc1 += "\n"
n_Doc1 += n_Doc2
 
with open ('Doc3.txt', 'w') as Doc3:
    Doc3.write(n_Doc1)
    
```
